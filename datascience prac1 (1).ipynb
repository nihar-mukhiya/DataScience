```python
from datascience import *
```

### creating table

```python
t = Table().with_columns([
        'letter', ['a', 'b', 'c', 'z'],
        'count',  [  9,   3,   3,   1],
        'points', [  1,   2,   2,  10],
    ])
```

```python
print(t)
```

### reading file and loading as Table

##### we use the function read_table() from class Table

```python
v = Table.read_table('galton.csv')
```

```python
print(v)
```

### reading csv's from url

```python
Table.read_table('http://data8.org/textbook/notebooks/sat2014.csv')
```

### accessing values

##### reading values of particular column

```python
v.column('mother')
```

```python
v['mother']
```

##### accessing values row wise

```python
v.row(0)
```

##### to find number of rows

```python
v.num_rows
```

### manipulating data

##### adding new column by method    with_column('label', 'values')

```python
t.with_column('money', [200, 100, 500, 1000])
```

##### selecting a column with select() method

```python
v.select('father')
```

##### renaming column with relabeled(label, new_label) method

```python
t.relabeled('letter', 'waiter')
```

##### accessing a row and displaying in table format using take() method

```python
t.take(1)
```

##### displaying rows that satisfy a particular constraint using where() method

###### where() method takes as input an array of boolean values

```python
t.where(t['points']<10)
```

```python
t.where('points', 2)
```

##### sorting columns using sort() method (by default in ascending)

```python
t.sort('count')
```

##### to sort in descending order

```python
t.sort('points', descending = True)
```

##### group() to group rows

```python
t.select(['count', 'points']).group('count', collect=sum)
```

### visualizing data

```python
import matplotlib
#matplotlib.use('Agg')
from datascience import Table
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

```python
normal_data = Table().with_columns([
'data1', np.random.normal(loc = 1, scale = 2, size = 100),
'data2', np.random.normal(loc = 4, scale = 3, size = 100)])
```

##### to display a histogram using hist() method

```python
normal_data.hist()
```

##### bins is used to divide the entire range of values into a series of intervals

```python
normal_data.hist(bins = range(-5, 10))
```

##### if we consider our table values as set of x,y points

##### plot()

```python
normal_data.plot()
```

##### scatter()

```python
# scatter() requires atleast one argument of a column compulsorily
# fit_line draws a line of best fit for each set of points
```

```python
normal_data.scatter('data2', fit_line = True)
```

##### barh() is used to display data in cateogrical way

```python
normal_data.barh('data1')
```
